```python
#Import necessary Packages
import numpy as np
import matplotlib.pyplot as plt
import h5py
import itertools
import scipy

#Import my own scripts
import bragg #bragg spectroscopy script
import spc #single photon counting script. Note my synthetic data generation is also in this script, out of convenience. 
```

```python
# Name of the hdf file that contain the data we need
f_name = 'sxro6416-r0504.h5'

# Open the hdf5 file, use the path to the images to extrate the data and place
# it in the image data object for further manipulation and inspection.
datafile = h5py.File(f_name, 'r')
image_data = []
for i in itertools.count(start=0):
    d = datafile.get(f'Configure:0000/Run:0000/CalibCycle:{i:04d}/Princeton::FrameV2/SxrEndstation.0:Princeton.0/data')
    if d is not None:
        # actual image is at first index
        image_data.append(d[0])
    else:
        break

# Tell me how many images were contained in the datafile
print(f"loaded {len(image_data)} images")
```

```python
calib_frame = image_data[0]
for i in image_data[1:]:
    calib_frame = calib_frame + i 
"""
For the image we use to calibrate all the other parameters we add up all of the frames to get better signal-to-noise ratio. 
In practice you can use any image that has good enough signal-noise ratio. 
"""

calib_frame = calib_frame/len(image_data)

%matplotlib inline
plt.figure(figsize=(16,14))
plt.imshow(calib_frame,cmap='flag')
plt.colorbar()
plt.show()
```

```python
x = [[],[]] #These will be the x-values of points on the two spectral lines
y = [[],[]] #These will be the y-values of points on the two spectral lines
```

```python
%matplotlib qt
#can't have matplotlib inline for this part because we are using matplotlib interactive mode
x[0],y[0] = bragg.find_spectral_line(calib_frame,80)
```

```python
%matplotlib qt
x[1],y[1] = bragg.find_spectral_line(calib_frame,80)
```

```python
#Here is a plot of the points we got for each line. 
%matplotlib inline
plt.figure(figsize=(4,3))
plt.imshow(calib_frame,cmap='flag')
plt.colorbar()
plt.plot(x[0],y[0],'*',color='lime',label='line 1 (1218.5 eV)')
plt.plot(x[1],y[1],'*',color='white',label='line 2 (1188.0 eV)')
plt.legend()
plt.show()
```

```python
#A closer look at the points for line 1 shows that one of those points is anomalous:
%matplotlib inline
plt.plot(y[0],x[0])
plt.xlabel('y coordinate')
plt.ylabel('x coordinate')
plt.title('Points found for line 1')
```

```python
#This function removes the furtherst outlier, and replaces it with the average of its two neighbours
x[0] = bragg.remove_outlier(x[0])
plt.plot(y[0],x[0])
plt.xlabel('y coordinate')
plt.ylabel('x coordinate')
plt.title('Points found for line 1')
```

```python
#We know that the experiment is set up to look at between 1100eV and 1600eV:
E1 = 1100 
E2 = 1600
E = [1218.5,1188.0] #If you selected the leftmost spectral line first then put this in decreasing order of energy. 
two_d = 15.96e-10 #2d for the crystal we are using 
dims = calib_frame.shape #dimensions of the image we're looking at 

```

```python
#Here we give a plot for the initial guess of what the parameters should be. 
initial_guess,theta = bragg.plot_initial_guess(calib_frame,E1,E2,x,y,E,two_d)
```

```python
#I have a loss function for least squares fitting later. 
#For reference we can look at how good this initial guess is:
residuals = bragg.residuals(initial_guess,x,y,theta)
loss = np.sum(np.square(residuals))
print(f"loss: {loss}")
```

```python
#The next step is to simply optimize those parameters. 
fitted_params = bragg.fit_params(initial_guess,x,y,theta,dims)



#Here I print those parameters to compare with above:
print("")
print(f"alpha: {fitted_params[0]}")
print(f"Pz: {fitted_params[1]}")
print(f"x0: {fitted_params[2]}")
print(f"y0: {fitted_params[3]}")
print(f"gamma: {fitted_params[4]}")
```

```python
bragg.plot_fitted_params(fitted_params,x,y,theta,dims)
```

```python
image = image_data[8] #We are looking at image 8 in this example
```

```python
#Now that we have all these parameters we can move on to the single photon counting part
size = image.shape
alpha = fitted_params[0]
Pz = fitted_params[1]
x0 = fitted_params[2]
y0 = fitted_params[3]
gamma = fitted_params[4]
N = 200 #Number of bins we want to use. 


energy_matrix,bins,pixels_per_bin,norm_to_au = spc.return_energy_matrix(two_d,size,x0,y0,gamma,alpha,Pz,N,return_normalisations=True)
#energy_matrix is a matrix that assigns an energy to each pixel, given the parameters. 
```

```python
#For a sanity check we can see an image of that energy matrix 
plt.figure(figsize=(16,16))
plt.imshow(energy_matrix,cmap='tab20')
plt.colorbar()
plt.title("Map of energies of pixels in the image (eV)")
#As you can see some bins have more pixels than others. We account for this later.
```

```python
#There are two ways to get a spectrum. Here we use single photon counting.
#This function chooses a threshold based on the distribution of ADU values in the image. 
#Then it looks for pixels above that threshold that are greater than in ADU value than all the pixels around it.
#This it counts as a photon hit. 
spc_spectrum = spc.return_spc(image,energy_matrix,bins)
```

```python
plt.figure(figsize=(16,10))
plt.plot(bins[3:185],spc_spectrum[3:185]/norm_to_au[3:185])
plt.xlabel("energy (eV)")
plt.ylabel("photons per pixel")
plt.title("Spectrum obtained from single-photon counting")
```

```python
np.sum(image)-np.sum(image-1)
```

```python
"""
The disadvantage of this method of single-photon counting is that 
1. it's not robust where there are many multi-photon events. 
    This is because when two photons overlap you may only have one local maximum to count as a photon
2. It's not robust when there's a very low number of photons 
    This is because there may be noise above the threshold level, and this will be counted. 
"""
```

```python
"""
The next way to obtain a spectrum, is by simply adding up the ADU values of all pixels in each bin
Advantage: In regimes of high and low density of photons, noise should even out and total ADU value should be linear with the number of photons
Disadvantage: You don't actually get a photon number.  
"""

ADU_spectrum = spc.sum_by_energy(energy_matrix,bins,image,normalise=False)
```

```python
plt.figure(figsize=(16,10))
ADU_per_pixel = ADU_spectrum/pixels_per_bin
plt.plot(bins[3:185],ADU_per_pixel[3:185]) #This method doesn't work very well on the edges so I chose these indices
plt.xlabel("energy (eV)")
plt.ylabel("average ADU count")
plt.title("Spectrum obtained from ADU counting")
#As we can see some of the features of this graph are significantly different to the graph above 
```

```python
"""
So in theory, if we assume that the single-photon counting works well in most regimes but not very well in some regimes,
and counting total ADU works well in all regimes but doesn't give a photon number or offset, we should be able to combine the two methods
This is useful because we can also also use this to give us an uncertainty. 
"""
%matplotlib qt
combined_spectrum, percent_uncertainty = spc.combine_spectra(ADU_spectrum,spc_spectrum,pixels_per_bin,bins,plot=True)

"""
As we can see in the graph, the two methods agree for the most part. However, in the regime with a large number of single-photon
events, the average ADU pixel will be very high but the number of photons counted will remain low. 

"""
```

```python
%matplotlib inline
plt.figure(figsize=(16,12))
i=5
j=180
spec = ADU_spectrum-60*pixels_per_bin
#plt.plot(bins[i:j],combined_spectrum[i:j]/norm_to_au[i:j])
error = percent_uncertainty[i:j]*combined_spectrum[i:j]/norm_to_au[i:j]
plt.errorbar(bins[i:j],combined_spectrum[i:j]/norm_to_au[i:j],error,ecolor='red')
plt.xlabel("Energy (eV)")
plt.ylabel("Intensity (au)")
```
